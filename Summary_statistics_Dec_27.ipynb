```python
import pandas as pd
adni_training=pd.read_csv("D:\\Research\\AD\\ADLLM\\ADNI process 9\\ADNI\\adni_thickness_Dec_17_training.csv")
adni_training=adni_training.fillna("NA")
```

```python
import pandas as pd
import numpy as np

# Example data
df = pd.DataFrame({
    'value':list(adni_training[adni_training["CENTILOIDS"]!="NA"]["CENTILOIDS"])
})

# Define custom bins with lower and upper bounds
bins = [(-np.inf, 20), (20, 40), (40, 60), (60, np.inf)]

# Create a new column for bin assignment
def assign_bin(value):
    for lower, upper in bins:
        if lower <= value < upper:
            return f"{lower} to {upper}"
    return None

df['bin'] = df['value'].apply(assign_bin)

stats_per_bin = df.groupby('bin')['value'].agg(['mean', 'std', 'count']).reset_index()
stats_per_bin
```
